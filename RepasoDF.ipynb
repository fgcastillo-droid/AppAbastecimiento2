```python
import pandas as pd
import numpy as np

# =====================================================================
# 1. CREACIÓN DE DATOS (DataFrame con índice por defecto 0, 1, 2...)
# =====================================================================
def cargar_datos_ejemplo() -> pd.DataFrame:
    """
    Crea un DataFrame simulado de inventario solar.
    NO tiene un índice específico (usa el 0, 1, 2... por defecto).
    Esto es exactamente lo que obtienes después de un reset_index().
    """
    datos = {
        'SKU': ['PAN-550W', 'INV-10KW', 'CAB-4MM', 'BAT-5KWH', 'EST-ALU'],
        'Categoria': ['Panel', 'Inversor', 'Cable', 'Bateria', 'Estructura'],
        'Stock': [150, 0, 1500, 10, 0],
        'Precio_USD': [120.0, 1500.0, 1.5, 2000.0, 45.0]
    }
    # No asignamos el SKU como índice. Queda plano.
    df = pd.DataFrame(datos)
    return df

# =====================================================================
# 2. LECTURA Y EXTRACCIÓN (Read) - ¡Aquí cambian las reglas!
# =====================================================================
def demostrar_extraccion_loc(df: pd.DataFrame):
    """Muestra cómo extraer datos cuando el índice es solo numérico."""
    print("--- 1. EXTRACCIÓN CON .loc (Índice Reseteado) ---")
    
    # A. Un valor exacto: Como 'PAN-550W' ya NO es el índice, 
    # tenemos que usar una condición para encontrar la fila.
    # .values[0] nos sirve para sacar el número limpio del resultado.
    precio_panel = df.loc[df['SKU'] == 'PAN-550W', 'Precio_USD'].values[0]
    print(f"Precio del Panel (Buscando por Condición): ${precio_panel}")
    
    # B. Varias filas específicas: Usamos .isin() en lugar de buscar por índice
    skus_deseados = ['PAN-550W', 'INV-10KW']
    cols_deseadas = ['SKU', 'Categoria', 'Stock'] # Agrego SKU para que se vea claro
    subset_especifico = df.loc[df['SKU'].isin(skus_deseados), cols_deseadas]
    print("\nSubset Específico (Usando .isin()):\n", subset_especifico)
    
    # C. Slicing (Rango) de filas: Ahora los "nombres" de las filas son 0, 1, 2...
    # ¡CUIDADO AQUÍ!: .loc incluye el último valor (el 2 SÍ se incluye).
    rango_filas = df.loc[0:2, :] 
    print("\nRango de Filas (Etiquetas del 0 al 2 inclusivo):\n", rango_filas)

# =====================================================================
# 3. FILTRADO CONDICIONAL (Search) - ¡Esto sigue funcionando idéntico!
# =====================================================================
def demostrar_filtros(df: pd.DataFrame):
    """
    El poder de las condiciones: No les importa si hiciste reset_index() o no.
    Funcionan exactamente igual.
    """
    print("\n--- 2. FILTRADO DE DATOS ---")
    
    # df.loc[ CONDICIÓN , [COLUMNAS] ]
    sin_stock_loc = df.loc[df['Stock'] == 0, ['SKU', 'Categoria', 'Precio_USD']]
    print("Sin Stock (usando .loc):\n", sin_stock_loc)
    
    # .query sigue siendo el rey de la legibilidad
    sin_stock_query = df.query("Stock == 0")[['SKU', 'Categoria', 'Precio_USD']]
    print("\nSin Stock (usando .query - Más limpio!):\n", sin_stock_query)

# =====================================================================
# 4. MODIFICACIÓN DE DATOS (Update) - ¡Sigue siendo la Regla de Oro!
# =====================================================================
def demostrar_modificacion_segura(df: pd.DataFrame) -> pd.DataFrame:
    """
    Incluso con un índice numérico, las modificaciones deben hacerse con .loc
    usando condiciones para encontrar las filas objetivo.
    """
    print("\n--- 3. MODIFICACIÓN SEGURA CON .loc ---")
    df_modificado = df.copy() 
    
    # CASO: Aplicar descuento del 10% SOLO a las Baterías.
    condicion_bateria = df_modificado['Categoria'] == 'Bateria'
    df_modificado.loc[condicion_bateria, 'Precio_USD'] = df_modificado['Precio_USD'] * 0.90
    
    # CASO: Crear nueva columna basada en stock
    df_modificado['Estado'] = 'Agotado'
    df_modificado.loc[df_modificado['Stock'] > 0, 'Estado'] = 'Activo'
    
    print("DataFrame Modificado:\n", df_modificado)
    return df_modificado

# =====================================================================
# 5. BÚSQUEDA POR POSICIÓN (iloc)
# =====================================================================
def demostrar_iloc(df: pd.DataFrame):
    """
    A .iloc no le importa qué hay en el índice. Siempre usa posiciones estrictas (0, 1, 2).
    A diferencia de .loc, en .iloc el último valor del rango (2) NO se incluye.
    """
    print("\n--- 4. BÚSQUEDA NUMÉRICA CON .iloc ---")
    
    # Traer desde la primera fila (0) hasta la 2 (sin incluir la 2) y columnas 0 a 2.
    primera_fila = df.iloc[0:2, 0:2]
    print("Filas 0 y 1, primeras 2 columnas:\n", primera_fila)

# =====================================================================
# EJECUCIÓN DEL SCRIPT
# ==================================================================
```

```python
df_inventario = cargar_datos_ejemplo()
df_inventario
```

```python
# 2. Ejecutamos las demostraciones
demostrar_extraccion_loc(df_inventario)
```

```python
demostrar_filtros(df_inventario)
```

```python
df_inventario_actualizado = demostrar_modificacion_segura(df_inventario)
```

```python
demostrar_iloc(df_inventario)
```
