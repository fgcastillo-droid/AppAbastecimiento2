# Pruebas de funciones

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from src.data.data_loader import _load_single_gsheet, _get_gsheet_client
from src.data.limpieza import limpiar_movimientos
from src import config
```

Buscamos la data de OINNM

```python
cliente = _get_gsheet_client()
id_sheet: str = config.GSHEET_IDS["OINM"] 

df_raw = _load_single_gsheet(client= cliente, sheet_id= id_sheet, sheet_name="OINM")
df_clean = limpiar_movimientos(df_raw)
```

```python
df_ab = df_clean[(df_clean.CodigoBodega.isin(["BF0001", "BF0004", "BF0008", "BF0009"])) ] 
```

# Salidas mensuales

```python
df_ab.groupby( [pd.Grouper(key='fecha', freq= 'ME')])['ValorMovimiento'].sum().sort_index(ascending=False)
```

# Salidas BF0009

```python
df_ab[(df_ab.CodigoBodega=='BF0009') & (df_clean.ValorMovimiento < 0)].groupby( [pd.Grouper(key='fecha', freq= 'ME')])['ValorMovimiento'].sum().sort_index(ascending=False)
```

# Salidas BF0008

```python
df_ab[(df_ab.CodigoBodega=='BF0008') & (df_clean.ValorMovimiento < 0)].groupby( [pd.Grouper(key='fecha', freq= 'ME')])['ValorMovimiento'].sum().sort_index(ascending=False)
```
